### Colunas de verificação de segmento da economia

```python
colunas_selecionadas = ['valor_Bruto_Agropecuaria', 'valor_Bruto_Industria', 'valor_Bruto_Servicos', 'valor_Bruto_Administracao_Publica', 'valor_Impostos_Liquido']
```

```python
indice_coluna_maior_valor = np.argmax(pib[colunas_selecionadas].values, axis=1)
```

```python
# Identificar coluna/segmento com maior valor
pib['melhor_Segmento'] = pd.Index(pib[colunas_selecionadas].stack().index[indice_coluna_maior_valor])
```

```python
# Identificar coluna/segmento com maior valor
pib['melhor_Segmento'] = list(pib['melhor_Segmento'].str)[1]
```

```python
# Identificar coluna/segmento com maior valor
pib['melhor_Segmento'] = pd.Index(pib[colunas_selecionadas].stack().index[indice_coluna_maior_valor])
```

```python
# Identificar coluna/segmento com maior valor
pib['melhor_Segmento'] = pib['melhor_Segmento'].str.slice(1, 2)
```

```python
# Identificar coluna/segmento com maior valor
pib['valor_melhor_Segmento'] = pib[colunas_selecionadas].apply(np.max, axis=1)
```
