### Importar dados do PIB

```python
pib.head()
```

### Realizar comando análogo função PROCV

#### PROCV entre PIB e estados

```python
pib = pd.merge(
    pib
    ,estado[['id_Estado', 'nome_Estado', 'id_Regiao']]
    ,left_on = ('id_Estado')
    ,right_on = ('id_Estado')
    ,how = 'inner'
)
```

#### PROCV entre PIB e regiões

```python
pib = pd.merge(
    pib
    ,regiao[['id_Regiao', 'sigla_Regiao']]
    ,left_on = ('id_Regiao')
    ,right_on = ('id_Regiao')
    ,how = 'inner'
)
```

```python
del pib['id_Regiao']
```

### Reoganizar posição da colunas

```python
pib = pib[[
    'nome_Estado'
    ,'sigla_Regiao'
    ,'id_Estado'
    ,'ano_PIB'
    ,'ano_Referencia'
    ,'valor_Bruto_Agropecuaria'
    ,'valor_Bruto_Industria'
    ,'valor_Bruto_Servicos'
    ,'valor_Bruto_Administracao_Publica'
    ,'valor_Bruto_Total'
    ,'valor_Impostos_Liquido'
    ,'valor_PIB'
    ,'valor_PIB_percapita'
    ,'populacao_Projetada'
    ,'quant_Municipios'
]]
```
